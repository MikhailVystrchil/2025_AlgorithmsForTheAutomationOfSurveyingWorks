### Задачки от нейросети

Нейросеть говорит, что все эти задачи можно решить осознав содержание этого видео, честно ответьте для себя так ли это?

> Может попробовать заняться этим на досуге?..

1. __Симуляция цепи Маркова: случайная прогулка__

Смоделируйте простую цепь Маркова с несколькими состояниями (например, два или три состояния: A, B, C), задайте вероятности переходов между ними и сгенерируйте последовательность длиной N. Выведите частоту появления каждого состояния.

2. __Перемешивание колоды (проверка случайности)__

Смоделируйте процесс перемешивания колоды из 5–10 карт: создайте алгоритм, который поочередно выбирает две карты и меняет их местами. После больших повторений визуализируйте, как часто каждая карта оказывается на первой позиции. Сколько перемешиваний нужно, чтобы получить равномерное распределение?

3. __Метод Монте-Карло для вероятности событий__

С помощью цепи Маркова смоделируйте выпадение орла/решки (монеты), чтобы оценить вероятность появления определенной последовательности, например, три орла подряд. Сравните экспериментальный результат с теоретическим.

4. __Предсказание следующего символа в тексте__

Задан короткий текст (например, “АБАБАБ”), постройте частотную таблицу переходов между буквами и напишите функцию, которая на основе текущей буквы случайным образом "предсказывает", какая буква может быть следующей.

5. __Оценка памяти у цепи Маркова__

Для вашей цепи Маркова проверьте, влияет ли на выбор следующего состояния только текущее, или можно добавить "историю": модифицируйте алгоритм так, чтобы следующий выбор зависел от прошлых двух состояний. Сравните результаты — что изменилось?

6. __Автодополнение по цепи Маркова__

Автоматически генерируйте короткий текст, используя цепь Маркова на основе заданной последовательности переходов (например, с вероятностями перехода между словами или буквами).

Каждая из этих задач строго основана на видео — все понятия и примеры в нем есть. Для реализации потребуется только базовый `Python` и возможно, библиотека `random` для случайного выбора.
