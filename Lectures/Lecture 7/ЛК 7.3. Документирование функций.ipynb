### –ó–∞—á–µ–º –Ω—É–∂–Ω—ã `docstring`?

‚úÖ –ü–æ–º–æ–≥–∞—é—Ç –ø–æ–Ω—è—Ç—å, —á—Ç–æ –¥–µ–ª–∞–µ—Ç —Ñ—É–Ω–∫—Ü–∏—è

‚úÖ –†–∞–±–æ—Ç–∞—é—Ç —Å `help()` –≤ –∏–Ω—Ç–µ—Ä–∞–∫—Ç–∏–≤–Ω–æ–º —Ä–µ–∂–∏–º–µ

```python
# –ü—Ä–∏–º–µ—Ä —Ñ—É–Ω–∫—Ü–∏–∏ –ë–ï–ó –¥–æ–∫—É–º–µ–Ω—Ç–∞—Ü–∏–∏
def add(a, b):
    return a + b

# –ü—Ä–∏–º–µ—Ä —Ñ—É–Ω–∫—Ü–∏–∏ –° –¥–æ–∫—É–º–µ–Ω—Ç–∞—Ü–∏–µ–π
def add_documented(a, b):
    """–°–ª–æ–∂–∏—Ç—å –¥–≤–∞ —á–∏—Å–ª–∞."""
    return a + b

# –ü—Ä–æ–≤–µ—Ä—è–µ–º __doc__
print(f"add.__doc__ = {add.__doc__}")
print(f"add_documented.__doc__ = {add_documented.__doc__}")

# –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ help()
print("\n--- help(add_documented) ---")
help(add_documented)
```

## 3. –û–¥–Ω–æ—Å—Ç—Ä–æ—á–Ω—ã–µ docstring'–∏

### –°–∏–Ω—Ç–∞–∫—Å–∏—Å –∏ –ø—Ä–∏–º–µ—Ä—ã

```python
# ‚úÖ –ü—Ä–∞–≤–∏–ª—å–Ω–æ: –æ–¥–Ω–æ—Å—Ç—Ä–æ—á–Ω–∞—è –¥–æ–∫—É–º–µ–Ω—Ç–∞—Ü–∏—è
def add(a: int, b: int) -> int:
    """–°–ª–æ–∂–∏—Ç—å –¥–≤–∞ —á–∏—Å–ª–∞."""
    return a + b

def greet(name: str) -> str:
    """–í–µ—Ä–Ω—É—Ç—å –ø—Ä–∏–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ–µ —Å–æ–æ–±—â–µ–Ω–∏–µ."""
    return f"–ü—Ä–∏–≤–µ—Ç, {name}!"

def is_positive(number: int) -> bool:
    """–ü—Ä–æ–≤–µ—Ä–∏—Ç—å, –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–µ –ª–∏ —á–∏—Å–ª–æ."""
    return number > 0

def to_upper(text: str) -> str:
    """–ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞—Ç—å —Ç–µ–∫—Å—Ç –≤ –≤–µ—Ä—Ö–Ω–∏–π —Ä–µ–≥–∏—Å—Ç—Ä."""
    return text.upper()

# –î–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏—è
print(f"add.__doc__ = {add.__doc__}")
print(f"greet.__doc__ = {greet.__doc__}")

# –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ help()
print("\n--- help(add) ---")
help(add)
```

### –ü–æ–≤–µ–ª–∏—Ç–µ–ª—å–Ω—ã–π —Ç–æ–Ω (–ø–æ–≤–µ–ª–∏—Ç–µ–ª—å–Ω–æ–µ –Ω–∞–∫–ª–æ–Ω–µ–Ω–∏–µ)

PEP 257 —Ä–µ–∫–æ–º–µ–Ω–¥—É–µ—Ç –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –ø–æ–≤–µ–ª–∏—Ç–µ–ª—å–Ω—ã–π —Ç–æ–Ω ‚Äî –∫–∞–∫ –±—É–¥—Ç–æ –æ—Ç–¥–∞—ë—Ç–µ –∫–æ–º–∞–Ω–¥—É:

```python
# ‚ùå –ù–µ–ø—Ä–∞–≤–∏–ª—å–Ω–æ: –æ–ø–∏—Å–∞—Ç–µ–ª—å–Ω—ã–π —Ç–æ–Ω
def bad_style(x):
    """–í–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∫–≤–∞–¥—Ä–∞—Ç —á–∏—Å–ª–∞."""
    return x ** 2

# ‚úÖ –ü—Ä–∞–≤–∏–ª—å–Ω–æ: –ø–æ–≤–µ–ª–∏—Ç–µ–ª—å–Ω—ã–π —Ç–æ–Ω
def good_style(x):
    """–í–µ—Ä–Ω—É—Ç—å –∫–≤–∞–¥—Ä–∞—Ç —á–∏—Å–ª–∞."""
    return x ** 2

# –î—Ä—É–≥–∏–µ –ø—Ä–∏–º–µ—Ä—ã
def calculate_sum(items):
    """–í—ã—á–∏—Å–ª–∏—Ç—å —Å—É–º–º—É –≤—Å–µ—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤."""
    return sum(items)

def validate_email(email):
    """–ü—Ä–æ–≤–µ—Ä–∏—Ç—å –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ—Å—Ç—å —ç–ª–µ–∫—Ç—Ä–æ–Ω–Ω–æ–π –ø–æ—á—Ç—ã."""
    return "@" in email

def remove_duplicates(items):
    """–£–¥–∞–ª–∏—Ç—å –¥—É–±–ª–∏–∫–∞—Ç—ã –∏–∑ —Å–ø–∏—Å–∫–∞."""
    return list(dict.fromkeys(items))

print(f"good_style.__doc__ = {good_style.__doc__}")
print(f"calculate_sum.__doc__ = {calculate_sum.__doc__}")
```

## 4. –ú–Ω–æ–≥–æ—Å—Ç—Ä–æ—á–Ω—ã–µ docstring'–∏

### –°—Ç—Ä—É–∫—Ç—É—Ä–∞ –º–Ω–æ–≥–æ—Å—Ç—Ä–æ—á–Ω–æ–≥–æ docstring'–∞

```python
def calculate_average(numbers):
    """
    –í—ã—á–∏—Å–ª–∏—Ç—å —Å—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ —Å–ø–∏—Å–∫–∞ —á–∏—Å–µ–ª.

    –≠—Ç–∞ —Ñ—É–Ω–∫—Ü–∏—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç —Å–ø–∏—Å–æ–∫ —á–∏—Å–ª–æ–≤—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∏—Ö
    –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ —Å—Ä–µ–¥–Ω–µ–µ. –ï—Å–ª–∏ —Å–ø–∏—Å–æ–∫ –ø—É—Å—Ç, –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç 0.

    –ê—Ä–≥—É–º–µ–Ω—Ç—ã:
        numbers: –°–ø–∏—Å–æ–∫ —á–∏—Å–ª–æ–≤—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π (—Ü–µ–ª—ã–µ –∏–ª–∏ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã–µ —á–∏—Å–ª–∞).

    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç:
        –°—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ —á–∏—Å–µ–ª –≤ –≤–∏–¥–µ —á–∏—Å–ª–∞ —Å –ø–ª–∞–≤–∞—é—â–µ–π —Ç–æ—á–∫–æ–π.
    """
    if not numbers:
        return 0
    return sum(numbers) / len(numbers)

# –ü—Ä–æ–≤–µ—Ä—è–µ–º docstring
print(calculate_average.__doc__)

# –¢–µ—Å—Ç–∏—Ä—É–µ–º
result = calculate_average([1, 2, 3, 4, 5])
print(f"\ncalculate_average([1, 2, 3, 4, 5]) = {result}")
```

### –ü–æ–¥—Ä–æ–±–Ω—ã–π –ø—Ä–∏–º–µ—Ä —Å –Ω–µ—Å–∫–æ–ª—å–∫–∏–º–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏

```python
def get_user_info(user_id: int, include_contacts: bool = False) -> dict:
    """
    –ü–æ–ª—É—á–∏—Ç—å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ –ø–æ –µ–≥–æ ID.

    –§—É–Ω–∫—Ü–∏—è –∏—â–µ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –≤ –±–∞–∑–µ –¥–∞–Ω–Ω—ã—Ö –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –µ–≥–æ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é.
    –ü—Ä–∏ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ—Å—Ç–∏ –º–æ–∂–Ω–æ –≤–∫–ª—é—á–∏—Ç—å –∫–æ–Ω—Ç–∞–∫—Ç–Ω—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é.

    –ê—Ä–≥—É–º–µ–Ω—Ç—ã:
        user_id: –£–Ω–∏–∫–∞–ª—å–Ω—ã–π –∏–¥–µ–Ω—Ç–∏—Ñ–∏–∫–∞—Ç–æ—Ä –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è.
        include_contacts: –ï—Å–ª–∏ True, –≤–∫–ª—é—á–∏—Ç—å –∫–æ–Ω—Ç–∞–∫—Ç–Ω—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é: False).

    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç:
        –°–ª–æ–≤–∞—Ä—å —Å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–µ–π –æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ –∏–ª–∏ –ø—É—Å—Ç–æ–π —Å–ª–æ–≤–∞—Ä—å, –µ—Å–ª–∏ –Ω–µ –Ω–∞–π–¥–µ–Ω.

    –ü—Ä–∏–º–µ—Ä –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è:
        >>> get_user_info(123)
        {'id': 123, 'name': 'Ivan', 'age': 30}
    """
    database = {
        123: {'name': 'Ivan', 'age': 30, 'email': 'ivan@example.com'},
        456: {'name': 'Maria', 'age': 28, 'email': 'maria@example.com'}
    }
    
    user = database.get(user_id, {})
    if user and not include_contacts:
        user = {k: v for k, v in user.items() if k != 'email'}
    return user

# –¢–µ—Å—Ç–∏—Ä—É–µ–º
result1 = get_user_info(123)
result2 = get_user_info(123, include_contacts=True)
print(f"–ë–µ–∑ –∫–æ–Ω—Ç–∞–∫—Ç–æ–≤: {result1}")
print(f"–° –∫–æ–Ω—Ç–∞–∫—Ç–∞–º–∏: {result2}")
```

## 5. Google style
### –û—Å–Ω–æ–≤–Ω—ã–µ —Å–µ–∫—Ü–∏–∏ Google style

1. **–ö—Ä–∞—Ç–∫–æ–µ –æ–ø–∏—Å–∞–Ω–∏–µ** ‚Äî –æ–¥–Ω–∞ —Å—Ç—Ä–æ–∫–∞
2. **–ü–æ–¥—Ä–æ–±–Ω–æ–µ –æ–ø–∏—Å–∞–Ω–∏–µ** ‚Äî –µ—Å–ª–∏ –Ω—É–∂–Ω–æ
3. **–ê—Ä–≥—É–º–µ–Ω—Ç—ã** (Args)
4. **–í–æ–∑–≤—Ä–∞—â–∞–µ–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ** (Returns)
5. **–ò—Å–∫–ª—é—á–µ–Ω–∏—è** (Raises)
6. **–ü—Ä–∏–º–µ—Ä—ã** (Examples)
7. **–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ** (Note)
8. **–ü—Ä–µ–¥—É–ø—Ä–µ–∂–¥–µ–Ω–∏–µ** (Warning)

```python
def divide(numerator: float, denominator: float) -> float:
    """
    –†–∞–∑–¥–µ–ª–∏—Ç—å –æ–¥–Ω–æ —á–∏—Å–ª–æ –Ω–∞ –¥—Ä—É–≥–æ–µ.

    –í—ã–ø–æ–ª–Ω—è–µ—Ç –¥–µ–ª–µ–Ω–∏–µ –¥–µ–ª–∏–º–æ–≥–æ –Ω–∞ –¥–µ–ª–∏—Ç–µ–ª—å —Å –æ–±—Ä–∞–±–æ—Ç–∫–æ–π –æ—à–∏–±–æ–∫.

    Args:
        numerator: –ß–∏—Å–ª–æ, –∫–æ—Ç–æ—Ä–æ–µ –Ω—É–∂–Ω–æ —Ä–∞–∑–¥–µ–ª–∏—Ç—å.
        denominator: –ß–∏—Å–ª–æ, –Ω–∞ –∫–æ—Ç–æ—Ä–æ–µ –¥–µ–ª–∏–º.

    Returns:
        –†–µ–∑—É–ª—å—Ç–∞—Ç –¥–µ–ª–µ–Ω–∏—è –≤ –≤–∏–¥–µ —á–∏—Å–ª–∞ —Å –ø–ª–∞–≤–∞—é—â–µ–π —Ç–æ—á–∫–æ–π.

    Raises:
        ValueError: –ï—Å–ª–∏ –¥–µ–ª–∏—Ç–µ–ª—å —Ä–∞–≤–µ–Ω –Ω—É–ª—é.
        TypeError: –ï—Å–ª–∏ –≤—Ö–æ–¥–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –Ω–µ —è–≤–ª—è—é—Ç—Å—è —á–∏—Å–ª–∞–º–∏.

    Examples:
        >>> divide(10, 2)
        5.0
        >>> divide(7, 2)
        3.5
    """
    if denominator == 0:
        raise ValueError("–î–µ–ª–∏—Ç–µ–ª—å –Ω–µ –º–æ–∂–µ—Ç –±—ã—Ç—å –Ω—É–ª–µ–≤—ã–º")
    return numerator / denominator

# –¢–µ—Å—Ç–∏—Ä—É–µ–º
print(f"divide(10, 2) = {divide(10, 2)}")
print(f"divide(7, 2) = {divide(7, 2)}")

# –ü–æ–∫–∞–∑—ã–≤–∞–µ–º docstring
print(f"\n--- Docstring ---")
print(divide.__doc__)
```

### –ü–æ–ª–Ω—ã–π –ø—Ä–∏–º–µ—Ä Google style —Å —Ç–∏–ø–∞–º–∏

```python
from typing import List

def process_data(data: List[int], multiplier: int = 2, skip_negative: bool = True) -> List[int]:
    """
    –û–±—Ä–∞–±–æ—Ç–∞—Ç—å —Å–ø–∏—Å–æ–∫ —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª —Å —Ñ–∏–ª—å—Ç—Ä–∞—Ü–∏–µ–π –∏ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ–º.

    –§—É–Ω–∫—Ü–∏—è –æ–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ—Ç —Å–ø–∏—Å–æ–∫ —á–∏—Å–µ–ª, –æ–ø—Ü–∏–æ–Ω–∞–ª—å–Ω–æ —Ñ–∏–ª—å—Ç—Ä—É—è –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–µ
    –∑–Ω–∞—á–µ–Ω–∏—è, –∞ –∑–∞—Ç–µ–º —É–º–Ω–æ–∂–∞–µ—Ç –∫–∞–∂–¥–æ–µ —á–∏—Å–ª–æ –Ω–∞ –∑–∞–¥–∞–Ω–Ω—ã–π –º–Ω–æ–∂–∏—Ç–µ–ª—å.

    Args:
        data: –°–ø–∏—Å–æ–∫ —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª –¥–ª—è –æ–±—Ä–∞–±–æ—Ç–∫–∏.
        multiplier: –ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –¥–ª—è —É–º–Ω–æ–∂–µ–Ω–∏—è –∫–∞–∂–¥–æ–≥–æ —á–∏—Å–ª–∞. –ü–æ —É–º–æ–ª—á–∞–Ω–∏—é: 2.
        skip_negative: –ï—Å–ª–∏ True, –ø—Ä–æ–ø—É—Å—Ç–∏—Ç—å –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–µ —á–∏—Å–ª–∞. –ü–æ —É–º–æ–ª—á–∞–Ω–∏—é: True.

    Returns:
        –°–ø–∏—Å–æ–∫ –æ–±—Ä–∞–±–æ—Ç–∞–Ω–Ω—ã—Ö —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª.

    Raises:
        TypeError: –ï—Å–ª–∏ –¥–∞–Ω–Ω—ã–µ –Ω–µ —è–≤–ª—è—é—Ç—Å—è —Å–ø–∏—Å–∫–æ–º –∏–ª–∏ –º–Ω–æ–∂–∏—Ç–µ–ª—å –Ω–µ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ.
        ValueError: –ï—Å–ª–∏ —Å–ø–∏—Å–æ–∫ –ø—É—Å—Ç.

    Note:
        –ò—Å—Ö–æ–¥–Ω—ã–π —Å–ø–∏—Å–æ–∫ –Ω–µ –∏–∑–º–µ–Ω—è–µ—Ç—Å—è.

    Warning:
        –ë–æ–ª—å—à–∏–µ –º–Ω–æ–∂–∏—Ç–µ–ª–∏ –º–æ–≥—É—Ç –ø—Ä–∏–≤–µ—Å—Ç–∏ –∫ –ø–µ—Ä–µ–ø–æ–ª–Ω–µ–Ω–∏—é —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª.

    Examples:
        >>> process_data([1, 2, -3, 4])
        [2, 4, 8]
        >>> process_data([1, 2, 3], multiplier=3)
        [3, 6, 9]
        >>> process_data([1, -2, 3], skip_negative=False)
        [2, -4, 6]
    """
    if not isinstance(data, list):
        raise TypeError("–î–∞–Ω–Ω—ã–µ –¥–æ–ª–∂–Ω—ã –±—ã—Ç—å —Å–ø–∏—Å–∫–æ–º")
    if not data:
        raise ValueError("–°–ø–∏—Å–æ–∫ –Ω–µ –º–æ–∂–µ—Ç –±—ã—Ç—å –ø—É—Å—Ç")

    filtered = data if not skip_negative else [x for x in data if x >= 0]
    return [x * multiplier for x in filtered]

# –¢–µ—Å—Ç–∏—Ä—É–µ–º
print(f"process_data([1, 2, -3, 4]) = {process_data([1, 2, -3, 4])}")
print(f"process_data([1, 2, 3], multiplier=3) = {process_data([1, 2, 3], multiplier=3)}")
print(f"process_data([1, -2, 3], skip_negative=False) = {process_data([1, -2, 3], skip_negative=False)}")
```

## 6. NumPy style

### –û—Å–Ω–æ–≤–Ω—ã–µ —Å–µ–∫—Ü–∏–∏ NumPy style

1. **–ö—Ä–∞—Ç–∫–æ–µ –æ–ø–∏—Å–∞–Ω–∏–µ**
2. **–†–∞—Å—à–∏—Ä–µ–Ω–Ω–æ–µ –æ–ø–∏—Å–∞–Ω–∏–µ**
3. **–ü–∞—Ä–∞–º–µ—Ç—Ä—ã** (Parameters)
4. **–í–æ–∑–≤—Ä–∞—â–∞–µ—Ç** (Returns)
5. **–í—ã–∑—ã–≤–∞–µ—Ç** (Raises)
6. **–°–º. —Ç–∞–∫–∂–µ** (See Also)
7. **–ü—Ä–∏–º–µ—á–∞–Ω–∏—è** (Notes)
8. **–ü—Ä–∏–º–µ—Ä—ã** (Examples)

```python
def calculate_statistics(data: List[float]) -> dict:
    """
    –í—ã—á–∏—Å–ª–∏—Ç—å –±–∞–∑–æ–≤—É—é —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É –¥–ª—è —Å–ø–∏—Å–∫–∞ —á–∏—Å–µ–ª.

    –†–∞—Å—à–∏—Ä–µ–Ω–Ω–æ–µ –æ–ø–∏—Å–∞–Ω–∏–µ: —ç—Ç–∞ —Ñ—É–Ω–∫—Ü–∏—è –∞–Ω–∞–ª–∏–∑–∏—Ä—É–µ—Ç —Å–ø–∏—Å–æ–∫ —á–∏—Å–ª–æ–≤—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π
    –∏ –≤—ã—á–∏—Å–ª—è–µ—Ç —Ä–∞–∑–ª–∏—á–Ω—ã–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏, –≤–∫–ª—é—á–∞—è —Å—Ä–µ–¥–Ω–µ–µ,
    –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏—è.

    Parameters
    ----------
    data : List[float]
        –°–ø–∏—Å–æ–∫ —á–∏—Å–ª–æ–≤—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞.

    Returns
    -------
    dict
        –°–ª–æ–≤–∞—Ä—å —Å–æ–¥–µ—Ä–∂–∏—Ç:
        - 'mean': –°—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ
        - 'min': –ú–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ
        - 'max': –ú–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ
        - 'count': –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤

    Raises
    ------
    ValueError
        –ï—Å–ª–∏ —Å–ø–∏—Å–æ–∫ –ø—É—Å—Ç.
    TypeError
        –ï—Å–ª–∏ —Å–ø–∏—Å–æ–∫ —Å–æ–¥–µ—Ä–∂–∏—Ç –Ω–µ-—á–∏—Å–ª–æ–≤—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è.

    See Also
    --------
    statistics.mean : –í—Å—Ç—Ä–æ–µ–Ω–Ω—ã–π –º–æ–¥—É–ª—å –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è —Å—Ä–µ–¥–Ω–µ–≥–æ
    numpy.mean : –§—É–Ω–∫—Ü–∏—è NumPy –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è —Å—Ä–µ–¥–Ω–µ–≥–æ

    Notes
    -----
    –≠—Ç–∞ —Ñ—É–Ω–∫—Ü–∏—è –∏—Å–ø–æ–ª—å–∑—É–µ—Ç –≤—Å—Ç—Ä–æ–µ–Ω–Ω—ã–µ —Ñ—É–Ω–∫—Ü–∏–∏ Python –±–µ–∑ –≤–Ω–µ—à–Ω–∏—Ö –±–∏–±–ª–∏–æ—Ç–µ–∫.

    Examples
    --------
    >>> data = [1, 2, 3, 4, 5]
    >>> stats = calculate_statistics(data)
    >>> stats['mean']
    3.0
    >>> stats['min']
    1
    """
    if not data:
        raise ValueError("–î–∞–Ω–Ω—ã–µ –Ω–µ –º–æ–≥—É—Ç –±—ã—Ç—å –ø—É—Å—Ç—ã")

    return {
        'mean': sum(data) / len(data),
        'min': min(data),
        'max': max(data),
        'count': len(data)
    }

# –¢–µ—Å—Ç–∏—Ä—É–µ–º
data = [1, 2, 3, 4, 5]
stats = calculate_statistics(data)
print("–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞:")
for key, value in stats.items():
    print(f"  {key}: {value}")
```

## 7. Sphinx style

### –û—Å–Ω–æ–≤–Ω—ã–µ –¥–∏—Ä–µ–∫—Ç–∏–≤—ã Sphinx

| –î–∏—Ä–µ–∫—Ç–∏–≤–∞ | –û–ø–∏—Å–∞–Ω–∏–µ |
|-----------|----------|
| `:param name:` | –û–ø–∏—Å–∞–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ |
| `:type name:` | –¢–∏–ø –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ |
| `:return:` | –û–ø–∏—Å–∞–Ω–∏–µ –≤–æ–∑–≤—Ä–∞—â–∞–µ–º–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è |
| `:rtype:` | –¢–∏–ø –≤–æ–∑–≤—Ä–∞—â–∞–µ–º–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è |
| `:raises ExceptionType:` | –û–ø–∏—Å–∞–Ω–∏–µ –∏—Å–∫–ª—é—á–µ–Ω–∏—è |
| `:note:` | –ü—Ä–∏–º–µ—á–∞–Ω–∏–µ |
| `:warning:` | –ü—Ä–µ–¥—É–ø—Ä–µ–∂–¥–µ–Ω–∏–µ |
| `:example:` | –ü—Ä–∏–º–µ—Ä –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è |

```python
def process_items(items: List[str], reverse: bool = False) -> List[str]:
    """
    –û–±—Ä–∞–±–æ—Ç–∞—Ç—å —Å–ø–∏—Å–æ–∫ —ç–ª–µ–º–µ–Ω—Ç–æ–≤.

    –°–æ—Ä—Ç–∏—Ä—É–µ—Ç —ç–ª–µ–º–µ–Ω—Ç—ã –∏ –æ–ø—Ü–∏–æ–Ω–∞–ª—å–Ω–æ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∏—Ö –≤ –æ–±—Ä–∞—Ç–Ω–æ–º –ø–æ—Ä—è–¥–∫–µ.

    :param items: –°–ø–∏—Å–æ–∫ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –¥–ª—è –æ–±—Ä–∞–±–æ—Ç–∫–∏
    :type items: List[str]
    :param reverse: –í–µ—Ä–Ω—É—Ç—å –ª–∏ —ç–ª–µ–º–µ–Ω—Ç—ã –≤ –æ–±—Ä–∞—Ç–Ω–æ–º –ø–æ—Ä—è–¥–∫–µ (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é: False)
    :type reverse: bool
    :return: –û—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫ —ç–ª–µ–º–µ–Ω—Ç–æ–≤
    :rtype: List[str]
    :raises ValueError: –ï—Å–ª–∏ —Å–ø–∏—Å–æ–∫ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –ø—É—Å—Ç
    :raises TypeError: –ï—Å–ª–∏ —ç–ª–µ–º–µ–Ω—Ç—ã –Ω–µ —è–≤–ª—è–µ—Ç—Å—è —Å–ø–∏—Å–∫–æ–º

    :note: –ò—Å—Ö–æ–¥–Ω—ã–π —Å–ø–∏—Å–æ–∫ –Ω–µ –∏–∑–º–µ–Ω—è–µ—Ç—Å—è
    :warning: –û—á–µ–Ω—å –±–æ–ª—å—à–∏–µ —Å–ø–∏—Å–∫–∏ –º–æ–≥—É—Ç –ø–æ—Ç—Ä–µ–±–∏—Ç—å –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–∞–º—è—Ç–∏

    Example:
        >>> process_items(['banana', 'apple', 'cherry'])
        ['apple', 'banana', 'cherry']
        >>> process_items(['banana', 'apple', 'cherry'], reverse=True)
        ['cherry', 'banana', 'apple']
    """
    if not isinstance(items, list):
        raise TypeError("–≠–ª–µ–º–µ–Ω—Ç—ã –¥–æ–ª–∂–Ω—ã –±—ã—Ç—å —Å–ø–∏—Å–∫–æ–º")
    if not items:
        raise ValueError("–°–ø–∏—Å–æ–∫ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –Ω–µ –º–æ–∂–µ—Ç –±—ã—Ç—å –ø—É—Å—Ç")

    result = sorted(items, reverse=reverse)
    return result

# –¢–µ—Å—Ç–∏—Ä—É–µ–º
result1 = process_items(['banana', 'apple', 'cherry'])
result2 = process_items(['banana', 'apple', 'cherry'], reverse=True)
print(f"process_items(['banana', 'apple', 'cherry']) = {result1}")
print(f"process_items(['banana', 'apple', 'cherry'], reverse=True) = {result2}")
```

### 2. –î–æ–∫—É–º–µ–Ω—Ç–∏—Ä—É–π—Ç–µ –≤—Å–µ –æ–±—â–µ—Å—Ç–≤–µ–Ω–Ω—ã–µ —Ñ—É–Ω–∫—Ü–∏–∏

```python
# ‚úÖ –•–æ—Ä–æ—à–æ: –≤—Å–µ —Ñ—É–Ω–∫—Ü–∏–∏ –∑–∞–¥–æ–∫—É–º–µ–Ω—Ç–∏—Ä–æ–≤–∞–Ω—ã

def public_function(x: int) -> int:
    """–û—Ç–∫—Ä—ã—Ç—ã–π API: –≤—ã—á–∏—Å–ª–∏—Ç—å –∫–≤–∞–¥—Ä–∞—Ç."""
    return _internal_calculation(x)

def _private_function(x: int) -> int:
    # –ü—Ä–∏–≤–∞—Ç–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –Ω–µ —Ç—Ä–µ–±—É–µ—Ç –¥–æ–∫—É–º–µ–Ω—Ç–∞—Ü–∏–∏
    return x * 2

def _internal_calculation(x: int) -> int:
    """–í–Ω—É—Ç—Ä–µ–Ω–Ω–∏–π —Ä–∞—Å—á–µ—Ç: —É–º–Ω–æ–∂–∏—Ç—å —á–∏—Å–ª–æ –Ω–∞ —Å–µ–±—è."""
    return x * x

print(f"public_function.__doc__ = {public_function.__doc__}")
print(f"_private_function.__doc__ = {_private_function.__doc__}")
```

### 3. –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ –ø–æ–≤–µ–ª–∏—Ç–µ–ª—å–Ω—ã–π —Ç–æ–Ω –≤ –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ

```python
# ‚ùå –ù–µ–ø—Ä–∞–≤–∏–ª—å–Ω–æ: –æ–ø–∏—Å–∞—Ç–µ–ª—å–Ω—ã–π —Ç–æ–Ω
def wrong1(x):
    """–í–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∫–≤–∞–¥—Ä–∞—Ç —á–∏—Å–ª–∞."""
    return x ** 2

# ‚ùå –ù–µ–ø—Ä–∞–≤–∏–ª—å–Ω–æ: —ç—Ç–æ –Ω–µ –æ–ø–∏—Å–∞–Ω–∏–µ
def wrong2(x):
    """–ö–≤–∞–¥—Ä–∞—Ç —á–∏—Å–ª–∞."""
    return x ** 2

# ‚úÖ –ü—Ä–∞–≤–∏–ª—å–Ω–æ: –ø–æ–≤–µ–ª–∏—Ç–µ–ª—å–Ω—ã–π —Ç–æ–Ω
def correct1(x):
    """–í–µ—Ä–Ω—É—Ç—å –∫–≤–∞–¥—Ä–∞—Ç —á–∏—Å–ª–∞."""
    return x ** 2

# ‚úÖ –ü—Ä–∞–≤–∏–ª—å–Ω–æ: –ø–æ–≤–µ–ª–∏—Ç–µ–ª—å–Ω—ã–π —Ç–æ–Ω
def correct2(x):
    """–í—ã—á–∏—Å–ª–∏—Ç—å –∫–≤–∞–¥—Ä–∞—Ç —á–∏—Å–ª–∞."""
    return x ** 2

print(f"–ù–µ–ø—Ä–∞–≤–∏–ª—å–Ω–æ: {wrong1.__doc__}")
print(f"–ü—Ä–∞–≤–∏–ª—å–Ω–æ: {correct1.__doc__}")
```

### 4. –î–æ–∫—É–º–µ–Ω—Ç–∏—Ä—É–π—Ç–µ –∏—Å–∫–ª—é—á–µ–Ω–∏—è

```python
def process_data(data: List[int], divisor: int) -> List[float]:
    """
    –û–±—Ä–∞–±–æ—Ç–∞—Ç—å –¥–∞–Ω–Ω—ã–µ –ø—É—Ç–µ–º –¥–µ–ª–µ–Ω–∏—è –∫–∞–∂–¥–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞.

    Args:
        data: –°–ø–∏—Å–æ–∫ —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª –¥–ª—è –æ–±—Ä–∞–±–æ—Ç–∫–∏.
        divisor: –î–µ–ª–∏—Ç–µ–ª—å –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞.

    Returns:
        –°–ø–∏—Å–æ–∫ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –¥–µ–ª–µ–Ω–∏—è.

    Raises:
        ValueError: –ï—Å–ª–∏ –¥–∞–Ω–Ω—ã–µ –ø—É—Å—Ç—ã.
        ZeroDivisionError: –ï—Å–ª–∏ –¥–µ–ª–∏—Ç–µ–ª—å —Ä–∞–≤–µ–Ω –Ω—É–ª—é.
        TypeError: –ï—Å–ª–∏ –¥–∞–Ω–Ω—ã–µ –Ω–µ —Å–ø–∏—Å–æ–∫ –∏–ª–∏ –¥–µ–ª–∏—Ç–µ–ª—å –Ω–µ —á–∏—Å–ª–æ–≤–æ–π.
    """
    if not data:
        raise ValueError("–î–∞–Ω–Ω—ã–µ –Ω–µ –º–æ–≥—É—Ç –±—ã—Ç—å –ø—É—Å—Ç—ã")
    if divisor == 0:
        raise ZeroDivisionError("–î–µ–ª–∏—Ç–µ–ª—å –Ω–µ –º–æ–∂–µ—Ç –±—ã—Ç—å –Ω—É–ª–µ–≤—ã–º")

    return [x / divisor for x in data]

# –¢–µ—Å—Ç–∏—Ä—É–µ–º
try:
    result = process_data([10, 20, 30], 5)
    print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç: {result}")
except Exception as e:
    print(f"–û—à–∏–±–∫–∞: {e}")
```

### 5. –í–∫–ª—é—á–∞–π—Ç–µ –ø—Ä–∏–º–µ—Ä—ã –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è

```python
def factorial(n: int) -> int:
    """
    –í—ã—á–∏—Å–ª–∏—Ç—å —Ñ–∞–∫—Ç–æ—Ä–∏–∞–ª —á–∏—Å–ª–∞ n.

    Args:
        n: –ù–µ–æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–µ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ.

    Returns:
        –§–∞–∫—Ç–æ—Ä–∏–∞–ª —á–∏—Å–ª–∞ n.

    Raises:
        ValueError: –ï—Å–ª–∏ n –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–µ.

    Examples:
        >>> factorial(0)
        1
        >>> factorial(5)
        120
        >>> factorial(10)
        3628800
    """
    if n < 0:
        raise ValueError("n –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å –Ω–µ–æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–º")
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

# –¢–µ—Å—Ç–∏—Ä—É–µ–º –ø—Ä–∏–º–µ—Ä—ã
print(f"factorial(0) = {factorial(0)}")
print(f"factorial(5) = {factorial(5)}")
print(f"factorial(10) = {factorial(10)}")
```

### 6. –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ doctest –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –ø—Ä–∏–º–µ—Ä–æ–≤

```python
import doctest

def add(a: int, b: int) -> int:
    """
    –°–ª–æ–∂–∏—Ç—å –¥–≤–∞ —á–∏—Å–ª–∞.

    Examples:
        >>> add(2, 3)
        5
        >>> add(-1, 1)
        0
        >>> add(10, 20)
        30
    """
    return a + b

# –ó–∞–ø—É—Å–∫–∞–µ–º doctest
results = doctest.testmod(verbose=True)
print(f"\n–†–µ–∑—É–ª—å—Ç–∞—Ç—ã: {results}")
```

```python
# –í–∞—Ä–∏–∞–Ω—Ç 1: Google style (–ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –µ–≥–æ)
def google_function_1(x: int) -> int:
    """–í—ã—á–∏—Å–ª–∏—Ç—å –∫–≤–∞–¥—Ä–∞—Ç —á–∏—Å–ª–∞.

    Args:
        x: –í—Ö–æ–¥–Ω–æ–µ —á–∏—Å–ª–æ.

    Returns:
        –ö–≤–∞–¥—Ä–∞—Ç —á–∏—Å–ª–∞ x.
    """
    return x ** 2

def google_function_2(x: int, y: int) -> int:
    """–°–ª–æ–∂–∏—Ç—å –¥–≤–∞ —á–∏—Å–ª–∞.

    Args:
        x: –ü–µ—Ä–≤–æ–µ —á–∏—Å–ª–æ.
        y: –í—Ç–æ—Ä–æ–µ —á–∏—Å–ª–æ.

    Returns:
        –°—É–º–º–∞ —á–∏—Å–µ–ª x –∏ y.
    """
    return x + y

print("‚úÖ –í—ã–±–∏—Ä–∞–π—Ç–µ –æ–¥–∏–Ω —Å—Ç–∏–ª—å –∏ –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ –µ–≥–æ –≤–µ–∑–¥–µ –≤ –ø—Ä–æ–µ–∫—Ç–µ")
print("\n–ü—Ä–∏–º–µ—Ä 1:")
print(google_function_1.__doc__)
print("\n–ü—Ä–∏–º–µ—Ä 2:")
print(google_function_2.__doc__)
```
